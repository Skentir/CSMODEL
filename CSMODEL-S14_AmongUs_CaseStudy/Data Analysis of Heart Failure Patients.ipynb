## 1. Data Description

The dataset used in this case study is based on the medical records of heart failure patients collected at the Faisalabad Institute of Cardiology and at the Allied Hospital in Faisalabad (Punjab, Pakistan) from April 2015 to December 2015 [1, 2]. The researchers of the original study ensured that all patients who participated in the data collection are aware of the situation and have provided their consent [2]. Additionally, the original study was approved by Institutional Review Board of Government College University, Faisalabad-Pakistan and followed the principles of Helsinki Declaration [2].

The dataset consisted of 299 observations, wherein one observation is a medical report of a heart failure patient. In total, there are 105 female patients and 194 male patients whose medical report was included in the dataset. All of the patients are above 40 years old and have left ventricular systolic dysfunction (LVSD), a complication caused when the left ventricle of the heart loses its ability to contract. The left ventricle of the heart is responsible for pumping oxygen-rich blood into circulation and LVSD results to the heart having weaker pumping force. This puts all the patients for this dataset under Class III or IV of New York Heart Association's (NYHA) Functional Classification – a classification for how limited a patient is during physical activity. Patients under Class III and IV are vulnerable to fatigue easily and experiences discomfort when any physical activity is performed. 

Based on Ahmad et. al's case study, the following features were considered as potential variables to explain mortality caused by cardiovascular diseases like heart failure: age, gender, blood pressure (BP), ejection fraction (EF), creatinine phosphokinase (CPK), platelet count, serum sodium, serum creatinine, and whether the patient has anemia, diabetes, and is a smoker [2]. 

Some information like anemia, high blood pressure, diabetes, sex, and smoking where represented as binary data – with 1 as an indicator that the feature is true for that particular patient and 0 amounting to false. For example, a patient who has 1 on anemia but 0 for high blood pressure means that the patient has anemia but no symptoms for high blood pressure. Anemia classifications were determined by the hospital physician using the patient's haematocrit level. If a patient has haematocrit levels less than 36%, the minimum for a normal haematocrit level, then they were considered as anemic. For the classification of high blood pressure and smoking status, the original researchers used physician reports. 

Numerical risk factors were generated from the patients' blood reports and medical records. EF pertains to the amount of blood pumped out by the left ventricle for each contraction and is represented in percentage format. CPK describes the level of CPK enzymes in the blood and is represented in micrograms per liter (mcg/L). CPK was used as a risk factor for high levels of CPK in the blood could indicate heart failure or injury because more CPK is flowing in the blood during muscle tissue damage. A risk factor related with CPK is serum creatinine. Serum creatinine is a waste product generated by creatinine that comes from the normal wear and tear on muscles of the body. Like CPK, high levels of serum creatinine could indicate heart failure. In the dataset, serum creatinine is represented in values of milligrams per deciliter (mg/dL). Serum sodium is the indicator used to measure the sodium level in a patient's blood. It is represented as milliequivalents per litre (mEq/L). Sodium is important for ensuring that muscles and nerves are functioning correctly. Therefore, abnormally low levels of sodium in the blood might be caused by a heart failure. Platelet count was recorded as kiloplatelets per millilitre (kiloplatelets/mL).

Other features of the dataset includes a follow-up period measured in days and a variable for whether a patient is still alive or not called "death event". Like the rest of binary data in the study, 1 indicates that a patient is already dead, while 0 is the indicator for patients that are still alive. Follow-up period refers to the act of monitoring a patient's health after treatment, including when a patient participates in a clinical study or trial for a period of time. In the study, follow-up time varies per patient and is the basis for the death event feature, meaning patients that died within the follow-up period will have their death event recorded as 1.

### 2.5. Exploratory Data Questions

> ### Question 1: Which clinical features in the dataset are most associated with deaths from heart failure?

To determine which clinical features are most assosiated with mortality from heart failure, a DataFrame containing only dead patients shall be created.

```python
deaths = heartfailure_df[heartfailure_df['DEATH_EVENT'] == 1]
```

The patients will then be grouped based on those who have a specific sypmtom/attribute and store the count in variables.

```python
# Anemia
d1 = deaths.groupby('anaemia')['DEATH_EVENT'].count()
# Diabetes
d2 = deaths.groupby('diabetes')['DEATH_EVENT'].count()
# High BP
d3 = deaths.groupby('high_blood_pressure')['DEATH_EVENT'].count()
# Smoking History
d4 = deaths.groupby('smoking')['DEATH_EVENT'].count()
# Sex
d5 = deaths.groupby('sex')['DEATH_EVENT'].count()
```

Next, the percentage of dead patients who had a specific symptom/attribute will be calculated.

```python
# Anemia
r1 = [round(x/sum(d1),4)*100 for x in d1]
# Diabetes
r2 = [round(x/sum(d2),4)*100 for x in d2]
# High BP
r3 = [round(x/sum(d3),4)*100 for x in d3]
# Smoking History
r4 = [round(x/sum(d4),4)*100 for x in d4]
# Sex
r5 = [round(x/sum(d5),4)*100 for x in d5]
```

Using the variables above, a new DataFrame will be created which counts the observations for dead and alive per clinical feature and percentage of the count.

```python
p1 = pd.DataFrame({"Count":[d1[0],d1[1]],"%":[r1[0],r1[1]]}, index=["Anemia: 0", "Anemia: 1"])
p2 = pd.DataFrame({"Count":[d2[0],d2[1]],"%":[r2[0],r2[1]]}, index=["Diabetes: 0", "Diabetes: 1"])
p3 = pd.DataFrame({"Count":[d3[0],d3[1]],"%":[r3[0],r3[1]]}, index=["High BP: 0", "High BP: 1"])
p4 = pd.DataFrame({"Count":[d4[0],d4[1]],"%":[r4[0],r4[1]]}, index=["Smoking: 0", "Smoking: 1"])
p5 = pd.DataFrame({"Count":[d5[0],d5[1]],"%":[r5[0],r5[1]]}, index=["Sex: 0", "Sex: 1"])
final = pd.concat([p1,p2,p3,p4,p5])
```

```python
final.style.set_caption("Death Count of Binary Features")
```

For numerical attributes, outliers will be dropped using Interquartile Range by calling the `drop_outliers` function.

```python
hf_func.drop_outliers(deaths, "age")
hf_func.drop_outliers(deaths, "platelets")
hf_func.drop_outliers(deaths, "serum_creatinine")
hf_func.drop_outliers(deaths, "serum_sodium")
hf_func.drop_outliers(deaths, "ejection_fraction")
hf_func.drop_outliers(deaths, "creatinine_phosphokinase")
```

The mean of the numerical attributes will be acquired and rounded off according to its significant figure.

```python
print("age: " + str(round(deaths["age"].mean(),0)))
print("platelets: " + str(round(deaths["platelets"].mean(),2)))
print("serum creatinine: " + str(round(deaths["serum_creatinine"].mean(),1)))
print("serum sodium: " + str(round(deaths["serum_sodium"].mean(),0)))
print("ejection fraction: " + str(round(deaths["ejection_fraction"].mean(),0)))
print("creatinine phosphokinase: " + str(round(deaths["creatinine_phosphokinase"].mean(),0)))
```

The prominent clinical features of patients who died from heart failure were men at the age of 66 who had a history of smoking.

Having an average of $242278.66$ kiloplatelets/mL, $1.5$ mg/dL of serum creatinine, $136$ mEq/L of serum sodium, $31\%$ of ejection fraction, and $340$ mcg/L of creatinine phosphokinase were prominent signs as well.

The follow are the parts of a boxplot:
1. **Bottom Black Horizontal Line** - current group's minimum: $(Q1 - 1.5 \times 0.25)$
2. **Box Bottom Line** - lower quartile (Q1) / 25th percentile of the current group.
3. **Box Middle Line** - median (Q2) / 50th percentile of the current group.
4. **Box Upper Line** - upper quartile (Q3) / 75th percentile of the current group.
5. **Top Black Horizontal Line** - current group's maximum value: $(Q3 + 1.5 \times 0.75)$
6. **Small Diamond Shape** - outlier data or erroneous data from the current group.

The researchers used KMeans clustering, a partitioning cluster algorithm, to segment the data and group them based on how similar it is to one another. The first step here is importing the scikit learn library that has KMeans and Silhouette Score included for a simpler clustering.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

Before starting the clustering, the data must be prepared first. Any unwanted columns are dropped.

```python
heart_new_1 = heartfailure_df.drop(['time','month number'], 1)
```

A common problem in KMeans clustering is deciding what the 'k' value is, in short, how many clusters should be created by the algorithm. The k value is a pre-requisite for KMeans and highly affects how distorted a cluster can be. The most optimal method used by researchers is the Silhouette method where a metric is used for clustering.

The Silhouette Score is calculated for each instance of a cluster:
$\text{Silhouette Coefficient} = \frac{(x-y)}{max(x,y)}$
where, $y$ is the mean intra cluster distance: mean distance to the other instances in the same cluster. $x$ depicts mean nearest cluster distance i.e. mean distance to the instances of the next closest cluster [5]. The range of the score is between -1 and 1, where being closer to 1 means the more optimal clustering.

The researchers now test between the k values of 2 to 9 and check which k value creates a cluster with a Silhouette score closer to 1.

```python
#Use silhouette score
range_n_clusters = list (range(2,10))
score_list = {}

for n in range_n_clusters:
    clusterer = KMeans(n_clusters=n)
    preds = clusterer.fit_predict(heart_new_1)
    centers = clusterer.cluster_centers_

    score = silhouette_score(heart_new_1, preds)
    print("At "+str(n)+" clusters, the score is: " + str(score))
    score_list.update({n:score})
```

```python
# Plot according to the n clusters (k) and its respective score
plt.plot(*zip(*sorted(score_list.items())))
plt.xlabel('Silhouette Score')
plt.ylabel('N Clusters')
plt.show()
```

The means of the alive and dead patients per feature were compared using **t-test** and the t-test was used whether the null hypothesis will be rejected or failed to be rejected. **`age`, `ejection_fraction`, `serum_creatinine`,** and **`serum_sodium`** rejected the null hypothesis using the $p-value$ and therefore showed that these features have a relationship with `DEATH_EVENT`. The testing was conducted in the previous section. `creatinine_phosphokinase` and `platelets` failed to reject the null hypothesis, pointing to the possibility that these features could possibly not have any relationship to `DEATH_EVENT`. 

One insight that can be garnered from this question is that although the null hypothesis was failed to be rejected, it does not mean that the null hypothesis is true. The researchers noticed that `serum_creatinine` is a significant feature but `creatinine_phosphokinase` is not. This forms a contradiction because both are related with the same enzyme [2]. The researchers believe that this could have been a **Type II Error**. One recommendation to resolve this is to repeat all the data analysis processes in a sample that has a bigger size. There are only 299 observations in the original dataset, which was then reduced to 297 observations after data cleaning. The researchers believe that the strength of statistical inference correctness can increase with a bigger sample size for this increases the chances to capture the differences in the statistical tests, as well as increasing the power of a test. Another method to avoid Type II Errors is to increase the significance level, meaning changing the $p-value$ from $0.05$ to $0.10$ or higher. However, this could further introduce more problems for changing the significance level increases the chance of a false positive which results to Type I Errors that falsely reject the null hypothesis when it is actually true.

### B. Which patient cluster would most likely contract *more* deaths and the *least* number of deaths because of heart failure?

The determined numerical features from Research Question 1, **`age`, `ejection_fraction`, `serum_creatinine`,** and **`serum_sodium`**, were used as the basis of finding the clusters. 

The statistical method used in determining the cluster is by using the $z-score$. Since each cluster provided the mean per feature, the total **mean** and **standard deviation** of all clusters per feature was used in finding the $z-score$. The testing for this question was conducted in the section above.

The formula that was applied to each observation of the important column features is as follows:

$$Z = \frac{x - 𝛍}{𝛔}$$

where $x$ is the original value, $𝛍$ is the mean of the feature, and $𝛔$ is the standard deviation of the feature. This formula was performed in each cluster to arrive at the $z-score$ $(Z)$.

Afterwards, the clusters where filtered based on what type of cluster was being found: **more** deaths or **least** deaths. The general criteria for filtering involves:
* the higher the age, the higher chance of mortality;
* the lower ejection fraction (EF) percentage, specifically lower than $40\%$, the higher chance of mortality;
* the higher creatinine acquired from serum creatinine tests, the higher chance of mortality; and
* the lower sodium acquired from serum sodium tests, the higher chance of mortality.

After performing all the operations, it was discovered that **Cluster $8$** is the cluster that has the most probability to contract **more** deaths due to heart failure and **Cluster $5$** is the cluster that has the most probability to contract **least** deaths due to heart failure. The statistical techniques and data modeling approach used in determining this conclusion is performed in the previous section.

One insight that can be garnered from this is that the features are every changing. As mentioned before, there are some numerical features that were not considered because they did not pass the **t-test** conducted. However, as the sample size increases and additionaly features are introduced in future data gathering, the accurateness of determining the cluster or even clusters would definitely increase.
